```python
#imports
import numpy as np
import pandas as pd

from glob import glob

import matplotlib.pyplot as plt
```

```python
#open datas
print("Chargement des données...")
results = np.load("evaluation/resultats (0.01to0.05_LR_and_GaussianNB).npz")['arr_0']
results = results[:2]
print("Done")
```

```python
print(results[0].T[0])

#we define the matrix score such that score[i] is the list of AUC score for the subject i+1 (with i between 0 and 11)
# for linear regression here

score = np.zeros((12,5))

for i in range(12):
    score[i] = results[0][i]
```

```python
%matplotlib inline

tr = np.arange(0.01,0.06,0.01)

for i in range(12):
    plt.plot(tr, score[i], 'x-', lw=2, label="subject "+str(i))

plt.axis([0.01, 0.05, 0.60, 0.90])
plt.title("AUC score for different training rate (Linear Regression)")
```

```python
print(results[0].T[0])

#we define the matrix score such that score[i] is the list of AUC score for the subject i+1 (with i between 0 and 11)
# for Gaussian Naive Bayesian here

score2 = np.zeros((12,5))

for i in range(12):
    score2[i] = results[1][i]
```

```python
tr = np.arange(0.01,0.06,0.01)

for i in range(12):
    plt.plot(tr, score2[i], 'x-', lw=2, label="subject "+str(i))

plt.axis([0.01, 0.05, 0.60, 0.90])
plt.title("AUC score for different training rate (Gaussian Naive Bayesian)")
```
